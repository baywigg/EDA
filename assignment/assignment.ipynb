**Q1.** In class, we talked about how to compute the sample mean of a variable $X$,
$$
m(X) = \dfrac{1}{N} \sum_{i=1}^N x_i
$$
and sample covariance of two variables $X$ and $Y$,
$$
\text{cov}(X,Y) = \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))(y_i - m(Y))).
$$
Recall, the sample variance of $X$ is
$$
s^2 = \dfrac{1}{N} \sum_{i=1}^N (x_i - m(X))^2.
$$
It can be very helpful to understand some basic properties of these statistics. If you want to write your calculations on a piece of paper, take a photo, and upload that to your GitHub repo, that's probably easiest.

# 1.
- We know that $m(X)$ can be written as:

$$
\dfrac{1}{N} \sum_{i=1}^{N} x_i
$$

- So, we can rewrite $m(a + bX)$ as :

$$
m(a + bX) = \dfrac{1}{N} \sum_{i=1}^{N} (a + b x_i)
$$

- We can now split this into:

$$
m(a + bX) = \dfrac{1}{N} \sum_{i=1}^{N} a + \dfrac{1}{N} \sum_{i=1}^{N} b x_i
$$

- We can factor out our constants:

$$
m(a + bX) = a \times \left( \dfrac{1}{N} \sum_{i=1}^{N} 1 \right) + b \times \left( \dfrac{1}{N} \sum_{i=1}^{N} x_i \right)
$$

- We can write the first term as so because the summation of ones up to $N$ is just $N$, and we can substitute $m(X)$ back in for the second term:

$$
m(a + bX) = a \times 1 + b \times m(X)
$$

- We have shown:

$$
m(a + bX) = a + b \times m(X).
$$


# 2. 
- We know that the sample covariance of $X$ and $Y$ is defined as:

  $$
  \text{cov}(X,Y) = \dfrac{1}{N} \sum_{i=1}^{N} (x_i - m(X))(y_i - m(Y))
  $$

- So, we can rewrite $\text{cov}(X, a + bY)$ as:

  $$
  \text{cov}(X, a + bY) = \dfrac{1}{N} \sum_{i=1}^{N} (x_i - m(X))((a + b y_i) - m(a + bY))
  $$

- We can now expand the second term inside the summation:

  $$
  \text{cov}(X, a + bY) = \dfrac{1}{N} \sum_{i=1}^{N} (x_i - m(X)) \left( a + b y_i - m(a + bY) \right)
  $$

- We know from our previous proof that:

  $$
  m(a + bY) = a + b m(Y)
  $$

- Substituting this in, we get:

  $$
  \text{cov}(X, a + bY) = \dfrac{1}{N} \sum_{i=1}^{N} (x_i - m(X)) \left( a + b y_i - (a + b m(Y)) \right)
  $$

- Since $ a $ is a constant, it cancels out:

  $$
  \text{cov}(X, a + bY) = \dfrac{1}{N} \sum_{i=1}^{N} (x_i - m(X)) \left( b (y_i - m(Y)) \right)
  $$

- We can factor $ b $ out of the summation:

  $$
  \text{cov}(X, a + bY) = b \times \dfrac{1}{N} \sum_{i=1}^{N} (x_i - m(X))(y_i - m(Y))
  $$

- Recognizing the remaining term as $\text{cov}(X, Y)$, we conclude:

  $$
  \text{cov}(X, a + bY) = b \times \text{cov}(X, Y).
  $$

# 3. 
- We know that the sample covariance of two variables $X$ and $Y$ is defined as:

  $$
  \text{cov}(X,Y) = \dfrac{1}{N} \sum_{i=1}^{N} (x_i - m(X))(y_i - m(Y))
  $$

- So, we can rewrite $\text{cov}(a + bX, a + bX)$ as:

  $$
  \text{cov}(a + bX, a + bX) = \dfrac{1}{N} \sum_{i=1}^{N} ((a + b x_i) - m(a + bX)) ((a + b x_i) - m(a + bX))
  $$

- From our previous proof, we know that:

  $$
  m(a + bX) = a + b m(X)
  $$

- Substituting this in:

  $$
  \text{cov}(a + bX, a + bX) = \dfrac{1}{N} \sum_{i=1}^{N} (a + b x_i - (a + b m(X))) (a + b x_i - (a + b m(X)))
  $$

- Since $ a $ is a constant, it cancels out:

  $$
  \text{cov}(a + bX, a + bX) = \dfrac{1}{N} \sum_{i=1}^{N} (b (x_i - m(X))) (b (x_i - m(X)))
  $$

- Factoring out $ b $ from both terms:

  $$
  \text{cov}(a + bX, a + bX) = b^2 \times \dfrac{1}{N} \sum_{i=1}^{N} (x_i - m(X)) (x_i - m(X))
  $$

- Recognizing the remaining sum as the definition of the sample variance:

  $$
  s^2 = \dfrac{1}{N} \sum_{i=1}^{N} (x_i - m(X))^2
  $$

- We conclude:

  $$
  \text{cov}(a + bX, a + bX) = b^2 s^2.
  $$

- In particular, since $\text{cov}(X,X) = s^2$, we have:

  $$
  \text{cov}(a + bX, a + bX) = b^2 \text{cov}(X,X).
  $$

A non-decreasing transformation preserves the order of the data, so if you take the median of $X$ and then apply the transformation, you get the same result as if you applied the transformation first and then found the median. This works for any quantile, not just the median. However, measures like the IQR or the overall range involve differences between quantiles or extreme values, and a non-linear transformation can distort these differences, so the same property does not hold for them unless the transformation is linear.

```python
# loading libraries
import pandas as pd
import scipy
import numpy as np
import seaborn as sns
```

```python
# loading the foreign gifts dataset
foreign_gifts = pd.read_csv("./data/ForeignGifts_edu.csv")
foreign_gifts.head()
```

```python
# Create histogram for the Gift Type column
foreign_gifts["Foreign Gift Amount"].hist()
# Print out description information about Gift Type
print(foreign_gifts["Foreign Gift Amount"].describe())
print("="*50)
# Print out the specific distribution of the counts of unique data in Gift Type
print(foreign_gifts["Foreign Gift Amount"].value_counts())
print("="*50)
# Check for any NA values
print(foreign_gifts["Foreign Gift Amount"].isna().sum())

# The Foreign Gift Amount histogram is right-skewed, with most values at the lower end and a few extreme outliers. The mean of 588,232.7 is much higher than the median of 94,615, indicating skewness. While most gifts are relatively small with the 25th-75th percentile being 5,700 to 376,142, a few large donations (up to $100M) significantly impact the distribution. The presence of negative values suggests refunds or corrections.
```

```python
# Create histogram for the Gift Type column
foreign_gifts["Gift Type"].hist()
# Print out description information about Gift Type
print(foreign_gifts["Gift Type"].describe())
print("="*50)
# Print out the specific distribution of the counts of unique data in Gift Type
print(foreign_gifts["Gift Type"].value_counts())
print("="*50)
# Check for any NA values
print(foreign_gifts["Gift Type"].isna().sum())

# For the Gift Type column, there are no null or missing values, with 3 unique values possible per row: Contract, Monetary Gift, or Real Estate. The most prevelant value is Contract, with Monetary Gift being behind by about 6,000, and Real Estate is by far the lowest with only 11 values in the 28,221 rows.
```

```python
# Kernel Density Plot for the log of Foreign Gift Amount

# Ensure all values are positive before applying log
foreign_gifts_gift_amount_positive = foreign_gifts[foreign_gifts["Foreign Gift Amount"] > 0]

# Show kernel density plot
sns.kdeplot(x=np.log(foreign_gifts_gift_amount_positive["Foreign Gift Amount"])).set(title="KD Plot for Logarithm of Foreign Gift Amount")
```

```python
# Kernel Density Plot for the log of Foreign Gift Amount conditional on gift type

fig = sns.kdeplot(data=foreign_gifts_gift_amount_positive, x="Foreign Gift Amount", hue="Gift Type", common_norm=False).set(title="KD Plot for Logarithm of Foreign Gift Amount Conditional on Gift Type")
```

```python
# The log-transformed Foreign Gift Amount shows a right-skewed distribution, indicating most gifts are small with fewer large ones. When conditioned on Gift Type, Monetary Gifts dominate lower amounts, while Contracts and Real Estate have broader distributions but lower densities. The distinct peaks suggest some gift types follow standardized amounts.
```

```python
# Get the value counts of the country of giftor, as each row represents a gift, then sort them from largest to smallest
top_countries_gifts = foreign_gifts["Country of Giftor"].value_counts().sort_values(ascending=False).reset_index()

# Display top 15 countries
top_countries_gifts.head(15)
```

```python
# Group the data by country, then extract just the foreign gift amount, sum it, then sort from high to low
top_countries_amount = foreign_gifts.groupby("Country of Giftor")["Foreign Gift Amount"].sum().sort_values(ascending=False).reset_index()

# Display top 15 countries
top_countries_amount.head(15)
```

```python
# Group the data by institution, then extract just the foreign gift amount, sum it, then sort from high to low
top_institutions_amount = foreign_gifts.groupby("Institution Name")["Foreign Gift Amount"].sum().sort_values(ascending=False).reset_index()

# Display top 15 institutions
top_institutions_amount.head(15)
```

```python
# Create histogram with seaborn, with the x value the name, and the y value the gift amount
fig = sns.barplot(x="Institution Name", y="Foreign Gift Amount", data=top_institutions_amount.head(15))
# rotate the labels so they do not overlap
fig.set_xticklabels(fig.get_xticklabels(), rotation=90)
```

```python
# Group the data by name, then extract just the foreign gift amount, sum it, then sort from high to low
giftors_most_money = foreign_gifts.groupby("Giftor Name")["Foreign Gift Amount"].sum().sort_values(ascending=False).reset_index()

giftors_most_money.head(15)
```
